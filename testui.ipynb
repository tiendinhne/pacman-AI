```python
import pygame
from map_loader import Maze
from pacman import Pacman
from utils import load_assets, draw_grid

# Khởi tạo pygame trong notebook
pygame.init()

# Load bản đồ
maze = Maze('maps/task02_pacman_example_map.txt')
assets = load_assets()
tile_size = 24

# Tạo màn hình ảo (không hiển thị)
surface = pygame.Surface((maze.width * tile_size, maze.height * tile_size))

# Vẽ mê cung lên surface
draw_grid(surface, maze, assets, tile_size)

# Lưu ảnh ra file để kiểm tra
pygame.image.save(surface, 'maze_preview.png')

# Hiển thị ảnh trong notebook
from IPython.display import Image
Image(filename='maze_preview.png')
```
